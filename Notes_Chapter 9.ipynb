# Concatenating multiple DataFrames together

```python
stocks_2016 = pd.read_csv("data/stocks_2016.csv", index_col="Symbol")
stocks_2017 = pd.read_csv("data/stocks_2017.csv", index_col="Symbol")
stocks_2018 = pd.read_csv("data/stocks_2018.csv", index_col="Symbol")
```

```python
stocks_2016.head()
```

```python
stocks_2017.head()
```

```python
stocks_2018.head()
```

```python
pd.concat([stocks_2016,stocks_2017], keys = '2016 2017'.split(), names = 'Year Symbol'.split())
```

```python
pd.concat([stocks_2016,stocks_2017], keys = '2016 2017'.split(), names = 'Year Symbol'.split(), axis = 'columns')
```

```python
pd.concat([stocks_2016,stocks_2017],join = 'inner', keys = '2016 2017'.split(), names = 'Year Symbol'.split(), axis = 'columns')
#Default is inner
```

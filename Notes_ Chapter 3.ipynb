# Reducing memory by changing data types

```python
college = pd.read_csv('data/college.csv')
different_cols = ['RELAFFIL','SATMTMID','CURROPER','INSTNM','STABBR']
col2 = college.loc[:,different_cols]
col2.head()
# Not all columns can be coerced to the desired type. Take a look at the MENONLY column,
# which from the data dictionary appears to contain only 0/1 values. The actual data type of
# this column upon import unexpectedly turns out to be float64. The reason for this is that
# there happen to be missing values, denoted by np.nan. There is no integer representation
# for missing values. Any numeric column with even a single missing value must be a float.
```

```python
col2.dtypes
```

```python
original_memory = col2.memory_usage(deep = True)
original_memory
# For pandas to extract the exact amount of memory of an object data type column, the
# deep parameter must be set to True in the memory_usage method.
```

```python
col2['RELAFFIL'] = col2['RELAFFIL'].astype(np.int8)
col2.dtypes
```

```python
col2.memory_usage(deep=True)
```

```python
col2.select_dtypes(include = [object]).nunique() #number of unique values
```

```python
col2['STABBR']= col2['STABBR'].astype('category')
```

```python
print(col2.memory_usage(deep = True))
print(col2.dtypes)
```

```python
col2.memory_usage(deep=True)/original_memory
```

# Selecting the smallest of the largest

```python
movie = pd.read_csv("data/movie.csv")
movie2 = movie[['movie_title', 'imdb_score', 'budget']]
movie2.nlargest(100,'imdb_score') #Use the nlargest method to select the top 100 movies by imdb_score:
movie2.nlargest(100,'imdb_score').nsmallest(5,'budget') #Chain the nsmallest method to return the five lowest budget films among those
# with a top 100 score:
# It is possible to pass a list of column names to the columns parameter of the
# nlargest/nsmallest methods. This would only be useful to break ties in the event that
# there were duplicate values sharing the nth ranked spot in the first column in the list.
```

# Selecting the largest of each group by sorting

```python
movie = pd.read_csv('data/movie.csv')
movie2 = movie[['movie_title','imdb_score','title_year']]
```

```python
movie2.sort_values(['title_year','imdb_score'], ascending=False)
```

```python
movie_top_year = movie2.sort_values(['title_year','imdb_score'],ascending=False).drop_duplicates(subset=['title_year'])
movie_top_year
# By default, drop_duplicates keeps the very first appearance, but this behavior may be
# modified by passing the keep parameter last to select the last row of each group or False to
# drop all duplicates entirely.
```

```python
movie4 = movie[['movie_title', 'title_year','content_rating', 'budget']]
movie4_sorted = movie4.sort_values(['title_year','content_rating', 'budget'], ascending=[True,False,False])
movie4_sorted
```

# Replicating nlargest with sort_values

```python
movie_smallest_largest1 = (movie[['movie_title','imdb_score','budget']].nlargest(100,'imdb_score')
                          .sort_values(['budget']).head(5))
movie_smallest_largest2 = (movie[['movie_title','imdb_score','budget']].nlargest(100,'imdb_score')
                           .nsmallest(5,'budget'))
                          
movie_smallest_largest1.equals(movie_smallest_largest2)

# The issue arises because more than 100 movies exist with a rating of at least 8.4. Each of the
# methods, nlargest and sort_values, breaks ties differently, which results in a slightly
# different 100-row DataFrame.
```

```python
b=(movie[['movie_title','imdb_score','budget']].nlargest(100,'imdb_score'))
```

```python
a = (movie[['movie_title','imdb_score','budget']].sort_values('imdb_score').head(100))
```

```python
a.equals(b)
# Both above shoud have been same, but issue arises because more than 100 movies exist with a rating of at least 8.4. Each of the
# methods, nlargest and sort_values, breaks ties differently, which results in a slightly
# different 100-row DataFrame.
```

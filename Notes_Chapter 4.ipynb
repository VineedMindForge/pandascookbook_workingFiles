# Selecting data with both integers and labels

```python
col_start = college.columns.get_loc('UGDS_WHITE') #This will be printed
col_end = college.columns.get_loc('UGDS_UNKN') + 1 #+1 so will be printed
college.iloc[:5, col_start:col_end]
```

```python
# Selecting 10th till 15th inclusive
row_start = college.index[10]
row_end = college.index[15] #14th is the 15th, so 15 is correct
college.loc[row_start:row_end,'UGDS_WHITE':'UGDS_UNKN']
```

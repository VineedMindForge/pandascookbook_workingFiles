```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = 'python,c,c++,java'.split(',')
y = [90,60,7,85]
plt.bar(x,y, width=.5, color = 'green,yellow,blue,red'.split(","), align = 'edge', #align = center or edge
       edgecolor = 'red', linewidth=2, 
       linestyle =':', #Linestyle = ":"
       alpha = 1, label = "work")
plt.legend()
plt.xlabel("Language", fontsize = 8)
plt.ylabel("Popularity", fontsize = 8)
plt.title("Survey Report 2023", fontsize = 20)
plt.show()
```

# Filling values with unequal indexes

```python
baseball14 = pd.read_csv("data/baseball14.csv", index_col="playerID")
baseball15 = pd.read_csv("data/baseball15.csv", index_col="playerID")
baseball16 = pd.read_csv("data/baseball16.csv", index_col="playerID")
baseball14.head()
```

```python
baseball14.index.difference(baseball15.index)
```

```python
baseball15.index.difference(baseball14.index)
```

```python
hits_14 = baseball14['H']
hits_15 = baseball15['H']
hits_16 = baseball16['H']
print(len(hits_14+hits_15), len(hits_14), len(hits_15))
hits_14+hits_15
```

```python
hits_14.add(hits_15, fill_value=0)
```

```python
hits_total = hits_14.add(hits_15, fill_value=0) \
                    .add(hits_16, fill_value=0) 
hits_total
```

```python
hits_total.hasnans
```

# Appending columns from different DataFrames

```python
employee = pd.read_csv('data/employee.csv')
dept_sal = employee[['DEPARTMENT','BASE_SALARY']].sort_values(['DEPARTMENT','BASE_SALARY'], ascending = [True,False])
dept_sal
```

```python
max_dept_sal = dept_sal.drop_duplicates(subset='DEPARTMENT')
max_dept_sal.head()
```

```python
#max_dept_sal = max_dept_sal.set_index('DEPARTMENT')
employee = employee.set_index('DEPARTMENT')
```

```python
employee['MAX_DEPT_SALARY'] = max_dept_sal['BASE_SALARY']
```

```python
employee.head()
```

```python
employee[employee['BASE_SALARY']>employee['MAX_DEPT_SALARY']]
```

# Highlighting the maximum value from each column

```python
college = pd.read_csv('data/college.csv', index_col='INSTNM')
college.dtypes
```

```python
cols = ['MD_EARN_WNE_P10' , 'GRAD_DEBT_MDN_SUPP']
for col in cols:
    college[col] = pd.to_numeric(college[col],errors = 'coerce')
```

```python
college[cols]
```

```python
college_n = college.select_dtypes(include=np.number)
college.head()
```

```python
college.nunique()
```

```python
college.nunique()==2
```

```python
criteria = college.nunique()==2
criteria
```

```python
binary_cols = college.columns[criteria].tolist()
binary_cols
```

```python
college_n2 = college_n.drop(labels=binary_cols, axis=1)
college_n2.head()
```

```python
max_cols = college_n2.idxmax()
max_cols
```

```python
unique_max_cols = max_cols.unique()
unique_max_cols
```

```python
college_n2.loc[unique_max_cols].style.highlight_max()
```

# Replicating idxmax with method chaining

```python
college = pd.read_csv('data/college.csv', index_col='INSTNM')
cols = ['MD_EARN_WNE_P10', 'GRAD_DEBT_MDN_SUPP']
```

```python
for col in cols:
    college[col] = pd.to_numeric(college[col],errors = 'coerce')
```

```python
college_n = college.select_dtypes(include = np.number)
```

```python
criteria = college_n.nunique() ==2
criteria
```

```python
binary_cols = college_n.columns[criteria].tolist()
binary_cols
```

```python
college_n = college_n.drop(labels=binary_cols, axis='columns')    
```
